###**Handling missing data and giving sensible datatypes to columns**

```python
# Handling missing values in Income with median imputation.
superstore_data['Income'].fillna(superstore_data['Income'].median(), inplace=True)
```

```python
# Changing the datatype of Income to int as all salaries are mostly in int and 2 pennies after a dollar bill don't make any difference.
superstore_data['Income'] = superstore_data['Income'].astype(int)
```

```python
# Changing Dt_Customer to dates in the dataframe.
# errors='coerce'. This is there so that whereever date is not present, i.e. '########' is converted to NaT (Not a Time), a null value.
superstore_data['Dt_Customer'] = pd.to_datetime(superstore_data['Dt_Customer'],  errors = 'coerce')
```

```python
# Handling missing values in Dt_Customer with median imputation.
superstore_data['Dt_Customer'].fillna(superstore_data['Dt_Customer'].median(), inplace=True)
```

```python
# Retain only date component in Dt_Customer
superstore_data['Dt_Customer'] = superstore_data['Dt_Customer'].dt.date
```

```python
sns.histplot(superstore_data['Age'], kde = True)
```

```python
print(f"The expected value of age is {round(age_dist[0], 0)}")
```

```python
def encode_education(x):
    if x == 'Basic':
        return 0
    elif x == 'Graduation':
        return 1
    elif x == 'Master':
        return 2
    elif x == 'PhD':
        return 3
```

Finding the best value of min_samples

```python
best_score = -1
best_min_samples = None

for min_samples in range(1, 15):  # try a range
    # Estimate eps temporarily using k-distance if you want
    neigh = NearestNeighbors(n_neighbors=min_samples)
    distances, _ = neigh.fit(segmentation_data).kneighbors(segmentation_data)
    eps_guess = np.percentile(distances[:, -1], 90)  # rough estimate of eps

    # Add a small epsilon if eps_guess is 0 to avoid the error
    if eps_guess == 0:
        eps_guess = 1e-5 # or some other small value

    db = DBSCAN(eps=eps_guess, min_samples=min_samples).fit(segmentation_data)
    labels = db.labels_

    if len(set(labels)) <= 1 or (len(set(labels)) == 2 and -1 in set(labels)):
        continue  # skip all noise or only noise + 1 cluster

    score = silhouette_score(segmentation_data, labels)
    if score > best_score:
        best_score = score
        best_min_samples = min_samples

print("Best min_samples:", best_min_samples)
```

```python
neigh = NearestNeighbors(n_neighbors=best_min_samples)
nbrs = neigh.fit(segmentation_data)
distances, indices = nbrs.kneighbors(segmentation_data)

# Step 4: Take the distance to the k-th nearest neighbor for each point
k_distances = distances[:, -1]  # last column is the k-th neighbor distance

# Step 5: Sort distances and plot
k_distances = np.sort(k_distances)
plt.figure(figsize=(8, 4))
plt.plot(k_distances)
plt.xlabel('Data points (sorted)')
plt.ylabel(f'Distance to {best_min_samples}th nearest neighbor')
plt.title('k-distance Graph')
plt.grid(True)
plt.show()
```

We are getting an elbow in the graph at near eps=0.02.

```python
best_score = -1
best_params = {}

for eps in np.arange(0.02, 0.05, 0.01):
    min_samples = best_min_samples
    db = DBSCAN(eps=eps, min_samples=min_samples).fit(segmentation_data)
    labels = db.labels_

    # Skip cases where all points are labeled as noise
    if len(set(labels)) <= 1 or -1 in set(labels) and len(set(labels)) == 2:
        continue
    score = silhouette_score(segmentation_data, labels)
    if score > best_score:
        best_score = score
        best_params = {'eps': eps, 'min_samples': min_samples}

print("Best score:", best_score)
print("Best params:", best_params)
```
